## Membaca dataset play-tennis

```python
import pandas as p
```

```python
tennis = p.read_csv("./tennis.csv")
tennis
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
```

```python
tennis["outlook_"] = le.fit_transform(tennis["outlook"])
tennis["temp_"] = le.fit_transform(tennis["temp"])
tennis["humidity_"] = le.fit_transform(tennis["humidity"])
tennis["windy_"] = le.fit_transform(tennis["windy"])
tennis["play_"] = le.fit_transform(tennis["play"])
tennis
```

```python
x = tennis.iloc[:,5:-1].values
y = tennis.iloc[:,-1].values
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(
x, y, test_size = 0.1)
x_train
```

```python
from sklearn.naive_bayes import GaussianNB as NB
gnb = NB()
gnb.fit(x_train,y_train)
y_predict = gnb.predict(x_test)
y_predict
```
